```python
from IPython.display import Image
import urllib2
import boto3
IMAGE_URL = (
    r'https://raw.githubusercontent.com/WillahScott/slack-pack/'
    r'f51c35bb7e7f37723e4b08ce2a9bd3e3e973aba9/data/grumpy_cat_2.png')

s3 = boto3.client('s3')

```

```python
s3.create_bucket(ACL='public-read', Bucket='awaybot_test')
```

```python
response = s3.list_buckets()

[i['Name'] for i in response['Buckets'] if 'Name' in i]
```

```python
response = urllib2.urlopen(IMAGE_URL)

s3.upload_fileobj(response, 'awaybot_test', 'test_pic')
```

```python
s3.download_file('awaybot_test', 'test_pic', 'test_pic.png')
```
